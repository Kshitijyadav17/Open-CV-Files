# Cropping Image

```python
img_NG_rgb = cv2.imread(r"D:\Data science and Machine Learning\open cv\opencv_bootcamp_assets_NB2\New_Zealand_Boat.jpg",1)
plt.imshow(img_NG_rgb[:,:,::-1])
```

# Crop the image of boat portion

```python
img_crop = img_NG_rgb[200:400,350:550]
plt.imshow(img_crop[:,:,::-1])
```

# Resizing image

### Method 1 : Specifiying the fx and fy factor

```python
resized_croped_img_2x = cv2.resize(img_crop,None,fx=2,fy=2)
plt.imshow(resized_croped_img_2x[:,:,::-1])
```

```python
resized_croped_img_2x.shape
```

### Method: Specifiying excat size of image

```python
dim = (500,400)
resized_croped_img = cv2.resize(img_crop,dsize=dim)
plt.imshow(resized_croped_img[:,:,::-1])
```

# Flipping Images

```python
img_NZ_rgb = cv2.imread(r"D:\Data science and Machine Learning\open cv\opencv_bootcamp_assets_NB2\New_Zealand_Boat.jpg",1)
img_NZ_rgb = img_NZ_rgb[:,:,::-1]
```

```python
img_NZ_rgb_Horizontal = cv2.flip(img_NZ_rgb,flipCode=1) # Flip on y axis
img_NZ_rgb_Vertical = cv2.flip(img_NZ_rgb,flipCode=0) # Flip on x axis
img_NZ_rgb_both = cv2.flip(img_NZ_rgb,flipCode=-1) # Flip on both axis

plt.figure(figsize=(20,10))
plt.subplot(141);plt.imshow(img_NZ_rgb_Horizontal);plt.title("Horizontal flip")
plt.subplot(142);plt.imshow(img_NZ_rgb_Vertical);plt.title("Vertical flip")
plt.subplot(143);plt.imshow(img_NZ_rgb_both);plt.title("Both axis flip")
plt.subplot(144);plt.imshow(img_NZ_rgb);plt.title("Original Image")
```
